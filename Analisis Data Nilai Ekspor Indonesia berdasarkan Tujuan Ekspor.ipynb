## 1. Deskripsi Data Nilai Ekspor Indonesia

```python
#memanggil data ekspor melalui modul Pandas
import pandas as pd
data_ekspor = pd.read_csv('processed-nilai-ekspor-negara-tujuan-2005-2013.csv',delimiter = ',', encoding = "utf-8")
```

```python
#memunculkan data ekspor pada dataframe Pandas di notebook
data_ekspor
```

```python
#menghitung banyaknya jumlah data per kolom pada data ekspor
data_ekspor.count()
```

```python
#menyederhanakan tampilan data ekspor dengan melihat nilai total ekspor Indonesia ke semua Benua selama 2005-2013 
data_ekspor.groupby(['benua']).mean()['nilai_ekspor']
```

### Deskripsi Data
File Nilai Ekspor Negara Tujuan ini diperolehd dari www.go.id dengan format csv.
File ini memiliki 4428 baris dan 6 kolom informasi yang berisi tahun, bulan, benua, region, nama_negara, dan nilai_ekspor.

## 2. Statistik Data Nilai Ekspor Indonesia

```python
vol_eks = data_ekspor.nilai_ekspor.sum()
mean_eks = data_ekspor.nilai_ekspor.mean()
maks_eks = data_ekspor.nilai_ekspor.max()
min_eks = data_ekspor.nilai_ekspor.min()

print("Jumlah Nilai Ekspor 2005-2013 sebesar", vol_eks, "USD")
print("Rata-rata Nilai Ekspor 2005-2013 ialah", mean_eks, "USD")
print("Maksimum Nilai Ekspor 2005-2013 ialah", maks_eks, "USD")
print("Minimum Nilai Ekspor 2005-2013 ialah", min_eks, "USD")
```

```python
data_ekspor_13= data_ekspor.loc[data_ekspor['tahun']==2013]
vol_eks13 = data_ekspor_13.nilai_ekspor.sum()
mean_eks13 = data_ekspor_13.nilai_ekspor.mean()
maks_eks13 = data_ekspor_13.nilai_ekspor.max()
min_eks13 = data_ekspor_13.nilai_ekspor.min()

print("Jumlah Nilai Ekspor khusus tahun 2013 sebesar", vol_eks, "USD")
print("Rata-rata Nilai Ekspor khusus tahun 2013 sebesar", mean_eks, "USD")
print("Maksimum Nilai Ekspor khusus tahun 2013 sebesar", maks_eks, "USD")
print("Minimum Nilai Ekspor khusus tahun 2013 sebesar", min_eks, "USD")
```

## Visualisasi Data Nilai Ekspor Indonesia berdasarkan Tujuan

```python
#menyederhanakan tampilan data ekspor dengan melihat nilai ekspor per benua setiap tahunnya
data_ekspor.groupby(['tahun','benua']).mean()
```

```python
#Menampilkan total kontribusi nilai ekspor dari setiap benua dari tahun 2005-2013
ax = data_ekspor.groupby(['benua']).mean()['nilai_ekspor']
ax = ax.plot(kind='bar',title='Rata-Rata Nilai Ekspor Tahun 2005-2018 Berdasarkan Benua Tujuan',fontsize=14,figsize=(15,5))
ax.set_xlabel("Benua Tujuan", fontsize=18)
ax.set_ylabel("Nilai Ekspor (dalam USD)", fontsize=18)
```

```python
#menyederhanakan tampilan data ekspor dengan melihat nilai ekspor per benua khusus di tahun 2013
olah_ekspor_2013 = data_ekspor.loc[data_ekspor['tahun']==2013]
ax = olah_ekspor_2013.groupby(['benua']).mean()['nilai_ekspor']
ax = ax.plot(kind='bar',title='Rata-Rata Nilai Ekspor Tahun 2013 Berdasarkan Benua Tujuan',legend=True,fontsize=14,figsize=(15,5))
ax.set_xlabel("Benua Tujuan", fontsize=18)
ax.set_ylabel("Nilai Ekspor (dalam USD)", fontsize=18)
```

```python
#Melihat trend ekspor Indonesia ke benua Asia dan Timur Tengah mulai dari tahun 2005 hingga 2013
olah_ekspor_asia = data_ekspor.loc[data_ekspor['benua']=='Asia dan Timur Tengah']
ax = olah_ekspor_asia.groupby(['tahun']).mean()['nilai_ekspor']
ax = ax.plot(kind='line',title='Trend Nilai Ekspor ke Benua Asia dan Timur Tengah 2005-2013',legend=True,fontsize=12,figsize=(15,5))
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Ekspor (dalam USD)", fontsize=12)
ax.plot(figsize=())
```

```python
#Melihat komposisi negara terhadap Nilai Ekspor Indonesia di benua Asia dan Timur Tengah mulai dari tahun 2005 hingga 2013
olah_ekspor_negara_asia= data_ekspor.loc[data_ekspor['benua']=='Asia dan Timur Tengah']
ax = olah_ekspor_negara_asia.groupby(['nama_negara']).mean()['nilai_ekspor']
ax = ax.plot(kind='pie',title='Komposisi Nilai Ekspor Indonesia di Benua Asia dan Timur Tengah Berdasarkan Negara (2005-2013)',fontsize=18,figsize=(10,10),autopct='%.1f%%')
```

```python
#Melihat komposisi negara Jepang terhadap Nilai Ekspor Indonesia di benua Asia dan Timur Tengah mulai dari tahun 2005 hingga 2013
olah_ekspor_negara_jepang= data_ekspor.loc[data_ekspor['nama_negara']=='Jepang']
ax = olah_ekspor_negara_jepang.groupby(['tahun']).mean()['nilai_ekspor']
ax = ax.plot(kind='line',title='Trend Nilai Ekspor Indonesia ke Negara Jepang 2005-2013',fontsize=12,figsize=(12,10))
```
